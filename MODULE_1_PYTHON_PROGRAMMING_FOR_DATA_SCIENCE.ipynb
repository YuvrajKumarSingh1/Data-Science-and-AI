```python
#encapsulation

#encap... is the technique of hiding an object's internal data and allowing controlled access through methods....



class BankAccount:

    def __init__(self, balance):
        self.__balance = balance      # Private Variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"₹{amount} Deposited Successfully")
        else:
            print("Invalid Amount")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"₹{amount} Withdraw Successfully")
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Current Balance:", self.__balance)


# Object Creation
obj = BankAccount(10000)

obj.deposit(5000)
obj.withdraw(2000)
obj.show_balance()
```

```python
#INHERITANCE

#parent class
class Animal:

  def sound(self):
    print("Animal make different sound")

#child class
class Dog(Animal):

      def bark(self):
        print("Dog says: Bark Bark!")

#object creation
dog = Dog()

dog.sound()  #Inheritance method
dog.bark()   #child class method
```
